```python
import requests
import time
from bs4 import BeautifulSoup

records=[]

for i in range (1,372):#372
    time.sleep(2) # 2 sec delay
    if i == 1:
        print(i)
        r = requests.get('https://www.bproperty.com/en/dhaka/apartments-for-sale/')
        soup = BeautifulSoup(r.text, 'html.parser')
        results = soup.find_all('li', attrs={'class':'ef447dde'}) 
        
    else:
        print(i)
        r= requests.get('https://www.bproperty.com/en/dhaka/apartments-for-sale/page-'+str(i)+'/')
        soup = BeautifulSoup(r.text, 'html.parser')
        results = soup.find_all('li', attrs={'class':'ef447dde'})  
    
    #records=[]
    for result in results:
        property_type= result.find('div', attrs={'aria-label':'Listing category'}).text
        location= result.find('div', attrs={'aria-label':'Listing location'}).text
        try:
            bed= result.find('span', attrs={'aria-label':'Beds'}).text
        except:
            bed = ""
        try:
            bath = result.find('span', attrs={'aria-label':'Baths'}).text
        except:
            bath=""
        
        total_sqft= result.find('span', attrs={'aria-label':'Area'}).text
        price= result.find('span', attrs={'aria-label':'Listing price'}).text
    
        records.append((property_type, location, bed, bath, total_sqft, price ))
    
    print(str(i)+": Completed: " +str(round(((i*100)/371),2))+"%")
    
 
```

```python
len(records)
```

```python
import pandas as pd
df = pd.DataFrame(records, columns=['property_type','location','bed','bath','total_sqft','price'])
```
